## Pandas
Pandas ek bahut powerful Python library hai jo data analysis ke liye use hota hai
Ye library data ko handle karne, process karne aur analyze karne me help karta hai

Main features of Pandas:
1. DataFrame banane ke liye - data ko table form me store karta hai
2. Data cleaning ke liye - missing values ko handle karta hai
3. Data filtering aur sorting ke liye
4. Different file formats (CSV, Excel, SQL) se data read/write kar sakte hai
5. Data visualization ke liye graphs bana sakte hai

# Example DataFrame banane ka:
```
python
import pandas as pd

data = {
    'Name': ['Raj', 'Priya', 'Amit'],
    'Age': [25, 30, 28],
    'City': ['Mumbai', 'Delhi', 'Pune']
```
