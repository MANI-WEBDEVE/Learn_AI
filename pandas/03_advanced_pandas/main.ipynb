## Merge the Data in Pandas

```python
products=pd.DataFrame({
    'ProductID': [1, 2, 3,4,5,6],
    'ProductName': ['Laptop', 'Phone', 'Book', "iphone", "tech", 'keyboard']
})
Salse=pd.DataFrame({
    'ProductID': [1, 2, 1, 3,1,2,1,1,1,1,5,6,4],
    'Sales': [50000, 30000, 40000, 500,50000, 30000, 50000,50000,50000,50000,45,666, "sold"]
})

merge_data=pd.merge(products, Salse, on="ProductID")
print(merge_data)

```

```python

merge_data=pd.merge(products, Salse, how='cross') # right , outer, inner, cross and left
merge_data
```

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
)
```

```python
data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]
},index=['a','b','h','i','j']
)
```

```python
data1.join(data2)
```

```python
data1.join(data2, how="inner")
```

```python
data1.join(data2, how="right")
```

```python
data1.join(data2, how="left")
```

```python
data1.join(data2,how='outer')
```
