```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

response = requests.get('https://www.olx.com.pk/houses_c1721?page=1',headers=headers)

```

```python

# Parse the response using BeautifulSoup with the lxml parser
soup = BeautifulSoup(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
title_web=soup.find_all('div', class_='_34bc0d5f _81d9cd00')

for i in title_web:
    print(i.text)
```

```python
location_web=soup.find_all('span', class_='f7d5e47e')

for i in location_web:
    print(i.text)
```

```python
card_web=soup.find_all('div', class_='_948d9e0a _8d76513c _371e9918')


for i in card_web:
    print(i.text.split(' ')[-1])
```

```python
info_house_web=soup.find_all('span', class_='_18b01e88')
print(info_house_web)
for i in info_house_web:
    print(i.text)
```

```python
house_big=soup.find_all("li", class_="undefined")

house_name=[]
house_price=[]
house_location=[]
house_info=[]
house_link=[]

for i in house_big:
    house_price.append(i.find("span", class_="ddc1b288").text.split(' ')[-1])
    house_name.append(i.find('h2', class_="_562a2db2").text)
    house_info.append(i.find('div', class_="_36ddd253").text)
    house_location.append(i.find("span", class_="f7d5e47e").text)
    house_link.append("https://www.olx.com.pk"+i.find("a")["href"])
```

```python
df=pd.DataFrame({'House Name': house_name, 'House Price': house_price, 'House Location': house_location, 'House Info': house_info, 'House Link': house_link})   
```

```python
final = pd.DataFrame()
for j in range(1, 11):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(f'https://www.olx.com.pk/houses_c1721?page={j}', headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    house_big = soup.find_all("li", class_="undefined")

    house_name = []
    house_price = []
    house_location = []
    house_info = []
    house_link = []

    for i in house_big:
        house_price.append(i.find("span", class_="ddc1b288").text.split(' ')[-1])
        house_name.append(i.find('h2', class_="_562a2db2").text)
        house_info.append(i.find('div', class_="_36ddd253").text)
        house_location.append(i.find("span", class_="f7d5e47e").text)
        house_link.append("https://www.olx.com.pk" + i.find("a")["href"])

    df = pd.DataFrame({
        'House Name': house_name,
        'House Price': house_price,
        'House Location': house_location,
        'House Info': house_info,
        'House Link': house_link
    })

    final = pd.concat([final, df], ignore_index=True)
```

```python
final.to_csv('olx_houses.csv', index=False)
```
