### Import the Importtant modules

```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X, Y=make_regression(n_samples=100, n_features=10, noise=0.1, random_state=40)
```

```python
X.shape
```

```python
Y.shape
```

### Split the data into traning and test data

```python
x_train, x_test, y_tran , y_test= train_test_split(X, Y, test_size=0.2, random_state=40)
```

```python
x_train.shape
```

## Train the Model for Train Dataset

```python
linear_reg=LinearRegression()
```

```python
linear_reg.fit(x_train, y_tran)
```

```python
y_preditc_value=linear_reg.predict(x_test)
```

```python
y_preditc_value
```

```python
y_test
```

```python
# mean squared error value
linear_reg_mse=mean_squared_error(y_test, y_preditc_value)
linear_reg_r2=r2_score(y_test, y_preditc_value)
print(
linear_reg_mse
)

print(linear_reg_r2)
```

## Ridge Regression

```python
ridge_reg=Ridge(alpha=0.1)
ridge_reg.fit(x_train, y_tran)
Y_prediction_reg=ridge_reg.predict(x_test)
ridge_reg_mse=mean_squared_error(y_test, Y_prediction_reg)
ridge_r2=r2_score(y_test, Y_prediction_reg)

print(ridge_reg_mse)
print(ridge_r2)
```
