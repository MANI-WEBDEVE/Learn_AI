```python
# visulize the data on graph format
plt.scatter(df["TV"], df["Sales"], c=df["TV"])
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("TV vs Sales")
plt.colorbar(label="TV")
plt.show()
```

```python
X=df[["TV"]]  # independent variable
Y=df["Sales"] # dependent variable
```

```python
Xtrain, Xtest, Ytrain,Ytest=train_test_split(X,Y,test_size=0.10, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit_transform(Xtrain)
```

```python
scaler.fit_transform(Xtest)
```

```python
model=LinearRegression()
model.fit(Xtrain,Ytrain)
```

```python
prediction=model.predict(Xtest)
# 'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.6, 9.7, 12.8, 25.5, 13.4]
prediction
```

```python
plt.scatter(Xtest, Ytest)
plt.plot(Xtest, prediction)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,root_mean_squared_error

mse=mean_squared_error(Ytest, prediction)
mae=mean_absolute_error(Ytest, prediction)
rmse=np.sqrt(mse)
rmse_1=root_mean_squared_error(Ytest, prediction)

mse,mae,rmse,rmse_1
```
