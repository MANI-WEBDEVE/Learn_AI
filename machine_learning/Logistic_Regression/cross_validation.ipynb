```python
# cross validation example dataset
# Stratified K-fold
import numpy as np
from sklearn.model_selection import StratifiedKFold
```

```python
X=np.random.rand(100,5)
```

```python
X
```

```python
y=np.array([0]*20 + [1]*80) # idhar + sign ka matalb two list ko combine karna ka hai
```

```python
y
```

```python
skf=StratifiedKFold(n_splits=5)
```

```python
skf
```

```python
for fold, (train_idx, test_idx) in enumerate(skf.split(X,y),1):
  print(f"fold {fold}")
  print("Train labels:", np.bincount(y[train_idx]))  # Count of class 0 and 1 in train
  print("Test labels: ", np.bincount(y[test_idx]))
```
