```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.data
```

```python
X=dataset.data
Y=dataset.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape
```

```python
tree_classifier=DecisionTreeClassifier(criterion="entropy")
```

```python
tree_classifier.fit(X_train,Y_train)
tree.plot_tree(tree_classifier, filled=True)
```

```python
# post pruning with max depth =2

tree_classifier_2=DecisionTreeClassifier(max_depth=2)
tree_classifier_2.fit(X_train,Y_train)
tree.plot_tree(tree_classifier_2, filled=True)

```

```python
Y_predict_1=tree_classifier.predict(X_test) # use criterion = entropy
Y_predict_2=tree_classifier_2.predict(X_test) # use criterion = gini b default

print(accuracy_score(Y_test,Y_predict_1))
print(accuracy_score(Y_test,Y_predict_2))
print("------------------------------------")
print(classification_report(Y_test,Y_predict_1))
print(classification_report(Y_test,Y_predict_2))
```

## Hyperparameter Tuning Prepruning

```python
parameter={
    'criterion':["gini","entropy"],
    'max_depth':[1,2,3,4,5],
    'splitter':["best","random"],
    'min_samples_leaf':[1,2,3,4,5],
    'min_samples_split':[2,3,4,5],
    'max_features':["sqrt","log2"] # 'auto' is deprecated in favor of 'sqrt'
}
```

```python
grid_tree_clf=GridSearchCV(DecisionTreeClassifier(),param_grid=parameter,cv=6, scoring="accuracy")
```

```python
grid_tree_clf.fit(X_train,Y_train)
```

```python
grid_tree_clf.best_params_
```

```python
Y_predict_grid=grid_tree_clf.predict(X_test)
print(accuracy_score(Y_test,Y_predict_grid))
print(classification_report(Y_test,Y_predict_grid))
```
